```python
#char
a = 0.5
b = 1
c = 2
d = 3
```

```python
df['gk_Shot_Stopper'] = (b*df.Speed + b*df.Stamina + d*df.GK_Positioning + d*df.GK_Diving + d*df.GK_Kicking + d*df.GK_Handling) / (2*b+ 4*d)
df['gk_skill'] = (b*df.Speed + b*df.Stamina + c*df.Long_Shots + c*df.Long_Shots + c*df.Short_Pass + + d*df.GK_Positioning + d*df.GK_Diving+ d*df.GK_Kicking + d*df.GK_Handling)/(2*b + 3*c + 4*d)
```

```python
plt.figure(figsize = (15,6))
sd1 = df.sort_values('gk_Shot_Stopper' ,ascending = False) [:5]
x1 = np.array(list(sd1['Name']))
y1 = np.array(list(sd1['gk_Shot_Stopper']))
x1 = x1
y1 = y1
sns.barplot(x1 ,y1 ,palette = 'colorblind')
plt.ylabel("shot stopping score")
```
